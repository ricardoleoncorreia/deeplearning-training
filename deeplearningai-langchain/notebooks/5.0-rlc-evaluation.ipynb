# LangChain: Evaluation

```python
from langchain.chains import RetrievalQA
from langchain_ollama import ChatOllama, OllamaEmbeddings
from langchain_community.document_loaders import WebBaseLoader
from langchain.vectorstores import DocArrayInMemorySearch
from langchain.indexes import VectorstoreIndexCreator
from IPython.display import display, Markdown
from langchain.prompts import ChatPromptTemplate
```

### Read web page

```python
loader = WebBaseLoader("https://www.automationexercise.com/")
docs = loader.load()
display(Markdown(docs[0].page_content))
```

### Create vector store

```python
embeddings = OllamaEmbeddings(model="all-minilm")

index = VectorstoreIndexCreator(
    vectorstore_cls=DocArrayInMemorySearch,
    embedding=embeddings
).from_loaders([loader])
```

### Create Q&A chain

```python
llm = ChatOllama(model='llama3.2:1b', temperature=0.0)

prompt_template = """
You are an expert in e-commerce websites. For each question, provide a concise and accurate answer
based on the provided context. If the answer is not found in the context, respond with "I don't know".
Context: {context}
Question: {question}
"""

prompt = ChatPromptTemplate.from_template(prompt_template)

qa = RetrievalQA.from_chain_type(
    llm=llm, 
    chain_type="stuff", 
    retriever=index.vectorstore.as_retriever(), 
    verbose=True,
    chain_type_kwargs = {
        "prompt": prompt,
        "document_separator": "<<<<>>>>>"
    }
)
```

### Create hard-coded Q&A examples

```python
hard_coded_examples = [
    {
        "query": "How much does a Winter Top cost?",
        "answer": "It costs Rs. 600"
    },
    {
        "query": "How many recommended items are there?",
        "answer": "There are 6 items: Stylish Dress, Winter Top, Summer White Top, Blue Top, Men Tshirt & Rs. 1000."
    }
]
```

### Generate examples by using LLMs

```python
from langchain.evaluation.qa import QAGenerateChain

example_gen_chain = QAGenerateChain.from_llm(llm)
generated_examples = example_gen_chain.invoke({"doc": docs[0]})
generated_examples["qa_pairs"]
```

### Combine examples

```python
examples = [*hard_coded_examples, generated_examples["qa_pairs"]]
examples
```

Make first demo query with first example

```python
qa.invoke(examples[2]["query"])
```

### LLM assisted evaluation

Get predictions for each example

```python
predictions = qa.batch(examples)
predictions
```

Create helper function to get the grade of the outputs

```python
import re

def extract_grade(text):
    """
    Extract CORRECT or INCORRECT from any part of a string
    
    Args:
        text (str): The string containing the grade
        
    Returns:
        str or None: 'CORRECT' or 'INCORRECT' if found, None otherwise
    """
    if not text:
        return None
    
    # Search for CORRECT or INCORRECT (case insensitive)
    match = re.search(r'\b(CORRECT|INCORRECT)\b', text, re.IGNORECASE)
    
    if match:
        return match.group(1).upper()
    else:
        return None
```

### Get graded outputs by using `QAEvalChain` 

```python
from langchain.evaluation.qa import QAEvalChain

eval_chain = QAEvalChain.from_llm(llm)

graded_outputs = eval_chain.evaluate(examples, predictions)
```

### Define helpers to track accuracy metrics

```python
correct_count = 0
total_count = len(graded_outputs)
grades = []
```

### Print output results

```python
for i, graded_output in enumerate(graded_outputs):
    print(f"Example {i+1}:")
    results_text = graded_output['results']
    grade = extract_grade(results_text)
    grades.append(grade)

    if grade == 'CORRECT':
        correct_count += 1

    print(f"Grade: {grade}")
    print(f"Full result: {results_text}")
    print()
```

### Calculate accurancy of the model

```python
accuracy = (correct_count / total_count) * 100 if total_count > 0 else 0

print(f"Accuracy: {accuracy:.1f}% ({correct_count}/{total_count})")
```
