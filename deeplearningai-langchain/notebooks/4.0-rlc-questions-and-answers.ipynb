# Q&A over Documents

An example might be a tool that would allow you to query a product catalog for items of interest.

```python
%%capture
%pip install beautifulsoup4 docarray
```

```python
from langchain.chains import RetrievalQA
from langchain_ollama import ChatOllama
from langchain_community.document_loaders import WebBaseLoader
from langchain.vectorstores import DocArrayInMemorySearch
from IPython.display import display, Markdown
from langchain.indexes import VectorstoreIndexCreator
from langchain_ollama import OllamaEmbeddings
```

Catalog will be downloaded from a webpage [Automation Exercise](https://www.automationexercise.com/) by using the `WebBaseLoader`

```python
loader = WebBaseLoader("https://www.automationexercise.com/")
docs = loader.load()
display(Markdown(docs[0].page_content))
```

Let's create the embedding class and the vectorstore

```python
embeddings = OllamaEmbeddings(model="all-minilm")

index = VectorstoreIndexCreator(
    vectorstore_cls=DocArrayInMemorySearch,
    embedding=embeddings
).from_loaders([loader])
```

Prepare the LLM:

```python
llm = ChatOllama(model='llama3.2:1b', temperature=0.0)
```

Define a query to be used in two examples:

```python
query ="Please list all the products shown in the page"
```

Query the vectorstore with the LLM and show the formatted response:

```python
response = index.query(query, llm=llm)
display(Markdown(response))
```

Let's try a couple of other examples:

```python
response = index.query("how many blue t-shirts are there?", llm=llm)
display(Markdown(response))
```

```python
response = index.query("please list all the tops in the product", llm=llm)
display(Markdown(response))
```

## Use of `RetrievalQA` to create the Q&A chain

Crate the in-memory storage:

```python
db = DocArrayInMemorySearch.from_documents(docs, embeddings)
retriever = db.as_retriever()
```

Generate the Q&A chain:

```python
qa_stuff = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever, verbose=True)
```

Query the chain:

```python
response = qa_stuff.invoke(query)
display(Markdown(response["result"]))
```
